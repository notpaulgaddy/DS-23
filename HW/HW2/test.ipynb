```python
class Card:
  deck = []  #static field, shared by class

  def __init__(self, suit, pv, rank):
    self.suit = suit
    self.rank = rank
    self.pv = pv

    if len(Card.deck) == 0:
      for i in range(0,52):
        Card.deck.append(False)


  def __str__(self):
    return (f"{self.rank} of {self.suit}, point value = {self.pv}")

  def __gt__(self, other):
    if self.pv > other.pv:
      return True
    return False

  def __ge__(self, other):
    if self.pv >= other.pv:
      return True
    return False

  def __lt__(self, other):
    if self.pv < other.pv:
      return True
    return False

  def __le__(self, other):
    if self.pv <= other.pv:
      return True
    return False

  def __eq__(self, other):
    if self.pv == other.pv:
      return True
    return False

  def __ne__(self, other):
    if self.pv != other.pv:
      return True
    return False

  def addToDeck(self):
    if self.suit=="Clubs":
      Card.deck[self.pv -2] = True
    elif self.suit=="Diamonds":
      Card.deck[self.pv -2 + 13] = True
    elif self.suit=="Hearts":
      Card.deck[self.pv -2 + 26] = True
    else:
      Card.deck[self.pv -2 + 39] = True
```
